**Question-1: 
Suppose in ‘ipl18’, you want to filter out the teams that have an NRR greater than zero, and for which the ‘For’ score exceeds the ‘Against’ score, i.e. both the conditions should be satisfied. Which teams will be left after you perform the above filtration?**

a) CSK, MI
b) SRH,CSK, MI
c) SRH,CSK, RCB
d) SRK,CSK, MI,RCB

```python
ipl18[(ipl18["NRR"]>0)&(ipl18["For"]>ipl18["Against"])]
```

**Answer-1: b) SRH,CSK, MI**

**Question-2:
If all the stats are taken for both ‘ipl17’ and ‘ipl18’, which team with its total points greater than 25 will have the highest win percentage?**

```python
a=ipl17.append(ipl18)
a
```

```python
b=a[["Team","Won","Lost"]].groupby(["Team"]).sum()
b.sort_values(by=["Won"], inplace=True, ascending=False)
(b["Won"]/(b["Won"]+b["Lost"]))*100
c=b[b["Won"]+b["Lost"]>25]
result=(c["Won"]/(c["Won"]+c["Lost"]))*100
result
```

Answer-2: SRH has the highest win percentage with its total points greater than 25
